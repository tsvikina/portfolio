# Violinplot

Vraag: Hebben alle huizen een hogere huurprijs bij een betere staat (behalve in de stad Hyderabad)?

```python
#importeer de benodigde bibliotheken
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from google.colab import files
upload=files.upload()
```

```python
#load het csv.file in een DataFrame
df=pd.read_csv('House_Rent_Dataset.csv')
df.head()

#De verschillende states namen wijzigen en de df filtreren.
different_states={1: 'Worn down state', 2: 'Average state', 3: 'As new'}
df['State']=df['State'].replace(different_states)
filtered_df=df[df['State'] != 'As new']
```

```python
#De verschillende states namen wijzigen en de df filtreren.
plt.figure(figsize=(11,11))

violinplot1=sns.violinplot(data=filtered_df,x='City', y='Rent', hue='State', density_norm='count',split=True, hue_order=('Worn down state', 'Average state'))
violinplot1.set(ylabel='Monthly rent (INR)')
plt.title("The monthly rent of appartments in India, plotted against the city in which they are located, grouped by the houses's state")
plt.legend(loc='upper right')

plt.show()
```

__Stelling:__ Alle huizen hebben een hogere huurprijs bij een betere staat, behalve in de stad Hyderabad:

De meeste huizen die in betere staat zijn, hebben een hogere huurprijs. Er zijn echter ook uitzonderingen.

```python
#combined violin and swarm plot
plt.figure(figsize=(11,11))
combined_plot = sns.violinplot(data=filtered_df, x='City', y='Rent', hue='State', density_norm='count', split=True, hue_order=('Worn down state', 'Average state'))

sns.swarmplot(data=filtered_df, x='City', y='Rent', hue='State', dodge=True, size=0.7, palette='dark:black', ax=combined_plot, legend=False, hue_order=('Worn down state', 'Average state'))
combined_plot.set(ylabel='Monthly rent (INR)')

legend = plt.legend(title='State')
legend.set_bbox_to_anchor((1, 1))
plt.title("The monthly rent of appartments in India, plotted against the city in which they are located, grouped by the houses's state")
plt.show()
```


De informatie over Hyderabad wijkt opvallend af van de rest van de gegevens. Er lijkt een vergelijkbaar aantal gewone en verouderde huizen van de staat te zijn, die allebei voor vergelijkbare prijzen worden verhuurd.
